```python
import pandas as pd
import numpy as np
```

```python
# import dataset
df = pd.read_csv('food.csv')
print(f"shape: {df.shape}")
df.head()
```

```python
df.info()
```

```python
# Check missing value
df.isnull().sum().to_frame('NaN value').T
```

```python
# check count of unique values in each columns
print("Unique values in each column:")
for col in df:
    print(f"{col}: {df[col].nunique()}")
```

```python
for col in df:
    if col == 'C_Type':
        print(f"{col}: {df[col].unique()}")
        print(f"{col}: {df[col].value_counts() if col == 'C_Type' else df[col].nunique()}")        
```
